Gradient

```python
import numpy as np
import matplotlib.pyplot as plt

# Função e seu gradiente
def f(x):
    return x**2

def grad_f(x):
    return 2*x

# Parâmetros iniciais
x_init = 8          # ponto inicial
lr = 0.1            # taxa de aprendizado (learning rate)
epochs = 10         # número de iterações

# Armazenar valores
x_vals = [x_init]
y_vals = [f(x_init)]

# Execução do Gradient Descent
x_current = x_init
for _ in range(epochs):
    grad = grad_f(x_current)
    x_current = x_current - lr * grad
    x_vals.append(x_current)
    y_vals.append(f(x_current))

# Visualização
x_range = np.linspace(-10, 10, 400)
plt.figure(figsize=(10, 6))
plt.plot(x_range, f(x_range), label='f(x) = x²', color='blue')
plt.scatter(x_vals, y_vals, color='red')
plt.plot(x_vals, y_vals, linestyle='--', color='red', label='Caminho do Gradient Descent')
plt.title('Gradient Descent na função f(x) = x²')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
